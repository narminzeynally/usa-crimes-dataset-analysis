> **filling null values of "Dispatch Date / Time"**

```python
usa_crimes_dataset[usa_crimes_dataset['Dispatch Date / Time'].isnull()][['Incident ID', 'Dispatch Date / Time', 'Start_Date_Time']]
```

```python
time_diff = pd.to_datetime(usa_crimes_dataset['Dispatch Date / Time']) - pd.to_datetime(usa_crimes_dataset['Start_Date_Time'])

mean_time_diff = time_diff.mean()
mode_time_diff = time_diff.mode()[0]

print(f'Mean time difference = {mean_time_diff}')
print(f'Mode time difference = {mode_time_diff}')

usa_crimes_dataset['Dispatch Date / Time'].fillna(pd.to_datetime(usa_crimes_dataset['Start_Date_Time']) + mode_time_diff, inplace=True)
```

> **filling of null values of "city" column according to "zip code" column**



```python
null_city_rows = usa_crimes_dataset[usa_crimes_dataset['City'].isnull()]
null_city_rows[['City', 'Zip Code']]
```

```python
zip_to_city = usa_crimes_dataset[usa_crimes_dataset['City'].notnull()].drop_duplicates('Zip Code').set_index('Zip Code')['City']
usa_crimes_dataset['City'] = usa_crimes_dataset['City'].fillna(usa_crimes_dataset['Zip Code'].map(zip_to_city))
```

```python
null_city_rows = usa_crimes_dataset[usa_crimes_dataset['City'].isnull()]
null_city_rows
```

> **deleting of columns that they have many null values and we don't need during analysis**

```python
dropping_columns = ['Street Prefix', 'Street Suffix', 'End_Date_Time', 'Block Address', 'Address Number', 'PRA']
usa_crimes_dataset = usa_crimes_dataset.drop(columns = dropping_columns)
```

```python
usa_crimes_dataset.isna().sum()
```

> **deleting rows that "latitude", "longitude" columns have "0" value**

```python
columns_have_zero = ['Latitude', 'Longitude']

for column in columns_have_zero:
  usa_crimes_dataset = usa_crimes_dataset.drop(usa_crimes_dataset[usa_crimes_dataset[column] == 0].index)
```

> **deleting rows that "city" column have digit values**

```python
columns_have_digits = ['City', 'State']

for column in columns_have_digits:
  usa_crimes_dataset = usa_crimes_dataset.drop(usa_crimes_dataset[(usa_crimes_dataset[column]).apply(lambda value: str(value).isdigit())].index)
```



> **replacing data in "Beat" column that has non-alphanumeric characters**

```python
usa_crimes_dataset['Beat'] = usa_crimes_dataset['Beat'].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

> **deleting rows that have null values in them**

```python
columns_have_null = ['Crime Name1', 'Police District Name', 'Zip Code', 'Beat', 'Street Name', 'Street Type']

for column in columns_have_null:
  usa_crimes_dataset = usa_crimes_dataset.dropna(subset = [column])
```

```python
usa_crimes_dataset.dtypes
```

```python
usa_crimes_dataset['Zip Code'] = (usa_crimes_dataset['Zip Code'].astype('int64')).astype('str')
```

> **adding id column**

```python
duplicate_counts = usa_crimes_dataset['Incident ID'].value_counts()
duplicates = duplicate_counts[duplicate_counts > 1]
duplicates
```

```python
usa_crimes_dataset.insert(0, 'id', [id for id in range(100001, len(usa_crimes_dataset) + 100001)])
```

> **seperating date and time columns**

```python
date_time_columns = {
    'Dispatch Date / Time' : ['Dispatch Date', 'Dispatch Time'],
    'Start_Date_Time' : ['Start Date', 'Start Time']
}

for column in date_time_columns:
  usa_crimes_dataset[column] = pd.to_datetime(usa_crimes_dataset[column], format='%m/%d/%Y %I:%M:%S %p')
  usa_crimes_dataset[date_time_columns[column][0]] = usa_crimes_dataset[column].dt.date
  usa_crimes_dataset[date_time_columns[column][1]] = usa_crimes_dataset[column].dt.time
  usa_crimes_dataset = usa_crimes_dataset.drop(column, axis=1)

usa_crimes_dataset
```

> **adding part of day according to time**

```python
def get_part_of_day(time):
  if (12 > int((str(time).split(':'))[0]) > 6):
    return 'Morning'
  elif (int((str(time).split(':'))[0]) == 6 and int((str(time).split(':'))[1]) == 0 and int((str(time).split(':'))[2]) == 0):
    return 'Morning'
  elif (int((str(time).split(':'))[0]) == 12 and int((str(time).split(':'))[1]) == 0 and int((str(time).split(':'))[2]) == 0):
    return 'Morning'
  elif (17 > int((str(time).split(':'))[0]) >= 12):
    return 'Afternoon'
  elif (int((str(time).split(':'))[0]) == 12 and int((str(time).split(':'))[1]) > 0 and int((str(time).split(':'))[2]) > 0):
    return 'Afternoon'
  elif (int((str(time).split(':'))[0]) == 17 and int((str(time).split(':'))[1]) == 0 and int((str(time).split(':'))[2]) == 0):
    return 'Afternoon'
  elif (20 > int((str(time).split(':'))[0]) >= 17):
    return 'Evening'
  elif (int((str(time).split(':'))[0]) == 17 and int((str(time).split(':'))[1]) > 0 and int((str(time).split(':'))[2]) > 0):
    return 'Evening'
  elif (int((str(time).split(':'))[0]) == 20 and int((str(time).split(':'))[1]) == 0 and int((str(time).split(':'))[2]) == 0):
    return 'Evening'
  else:
    return 'Night'


part_of_day = []
for index, start_time in enumerate(usa_crimes_dataset['start_time']):
  part_of_day.append(get_part_of_day(start_time))

usa_crimes_dataset.insert(len(usa_crimes_dataset.columns), 'start_part_of_day', part_of_day)

usa_crimes_dataset
```
